```python

# Make a hash table which will return the temperature and weather for a given Date and time slot
weatherMap = {}
folder_path1 = './training_data/training_data/Cleaned Data/Weather_Data'

for folder_path in os.listdir(folder_path1):
    file_path = os.path.join(folder_path1, folder_path)
    weather = pd.read_csv(file_path)
    

    # Iterate through each row of the weather dataframe
    for _, row in weather.iterrows():
        # Extract the date and time slot values from the row
        date1 = row['Date']
        time_slot = row['Time Slot']
        
        # Create a composite key using the date and time slot values
        key = (date1, time_slot)
        
        # Store the temperature and weather values in the weatherMap hash table
        weatherMap[key] = {'Temperature': row['Temperature'], 'Weather': row['Weather']}
        
def getAverageWeather(date):
    # Initialize the temperature and weather values
    temperature = 0
    weather = 0
    
    total = 0
    # Iterate through the 144 time slots
    for i in range(1, 144):
        # Create a composite key using the date and time slot values
        key = (date, i)
        
        # Add the temperature and weather values to the running total
        try:
            temperature += weatherMap[key]['Temperature']
            weather += weatherMap[key]['Weather']
            total += 1
        except:
            pass
        
    # Calculate the average temperature and weather values
    average_temperature = int(temperature // total)
    average_weather = int(weather // total)
    
    return {'Temperature': average_temperature, 'Weather': average_weather}
```

```python
import datetime

def extract_time(time_string):
    # Extract the time from the time string
    time = time_string.split(' ')[1]

    # Return the time
    return time
def extract_date(time_string):
    date = time_string.split(' ')[0]
    
    return date

def get_time_slot(time_string):
    # Define the start time
    start_time = datetime.datetime.strptime('00:00:00', '%H:%M:%S')

    # Define the time delta (10 minutes)
    delta = datetime.timedelta(minutes=10)

    # Convert the input time string to a datetime object
    input_time = datetime.datetime.strptime(time_string, '%H:%M:%S')

    # Loop over the time slots and find the matching time slot
    for i in range(1, 145):
        time_slot_start = start_time
        time_slot_end = start_time + delta
        if input_time >= time_slot_start and input_time < time_slot_end:
            return i
        start_time += delta

    # If the input time is not in any of the time slots, return None
    return None
```

```python

#This part of the code is to clean the Weather Data

folder_path = './training_data/training_data/weather_data'
for folder_path in os.listdir(folder_path):
    file_path = os.path.join(folder_path, folder_path)
    if os.path.isfile(file_path):
        tempfilename = folder_path + '/' + folder_path
        # filename+= '.csv'
        print(tempfilename)
        df = pd.read_csv(tempfilename,sep='\t',header=None)
        # Set the column names explicitly
        df.columns = ['Time Stamp', 'Weather', 'Temperature', 'PM 2.5']
        
        #add a column for the time slot
        df['Date'] = 0
        df['Time Slot'] = 0
        
        #Extract the time and date from the time stamp
        tempDate = df['Time Stamp'].apply(extract_date)
        df['Date'] = df['Time Stamp'].apply(extract_date)
        df['Time Slot'] = df['Time Stamp'].apply(extract_time)
        df['Time Slot'] = df['Time Slot'].apply(get_time_slot)
        
        #Calculate the average temprarture and weather and pm for each date and time slot
        df = df.groupby(['Time Slot']).mean()
        
        df = df.reset_index()
        df['Date'] = tempDate[0]
        
        outputPath = './training_data/training_data/Cleaned Data/Weather_Data/'+folder_path+'.csv'
        df.to_csv(outputPath, index=False)
```
